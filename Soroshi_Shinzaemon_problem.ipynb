# „ÄêProblem 2] Number of rice grains outside of the 100th day

```python
def compute_sorori_shinzaemon(n_days=100):
    list_n_grains = [];
    list_total_grains=[];
    total = 0
    for day in range(1, n_days + 1):
        grains_today = 2 ** (day - 1)
        total += grains_today
        list_n_grains.append(grains_today)
        list_total_grains.append(total)

    return list_n_grains, list_total_grains

list_n_grains, list_total_grains = compute_sorori_shinzaemon(n_days=334)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Output the total number of days you ake the course
print(f"Total grains of days i take the course: {list_total_grains[-1]}")

# Plot 
fig, ax = plt.subplots()
ax.plot(range(1,334+1), list_n_grains)

# Customize the plot
plt.title("Rice Grain Growth Over 100 Days",color='b')
plt.xlabel("Day [m]")
plt.ylabel("Number of Rice Grains [m]")
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.tick_params(labelsize=10)


# Show the plot
plt.show()
```

# „ÄêProblem 3] How many people can live for how many days with the rice we get?


```python
def calculate_survival_days(number_rice_grains, number_people):
  """
  I did a search on google with the question: What is the average consumption of rice per person?
  
  Calculates the number of days a group of people can survive on a given amount of rice.

  This is a simplified calculation and does not account for varying dietary needs, 
  activity levels, or other factors.

  Args:
    rice_grains: The total number of rice grains.
    num_people: The number of people in the group.

  Returns:
    The number of days the group can survive on the given amount of rice.
  """

  # Simplified assumption: 
  # - Average person consumes 100 grams of cooked rice per day
  # - 100 grams of cooked rice is approximately 50 grams of uncooked rice
  # - 1 gram of uncooked rice is roughly 25-30 grains (highly variable)
  # - Let's use an average of 27.5 grains per gram

  grains_per_person_per_day = 50 * 27.5  # 50 grams * 27.5 grains/gram 
  total_consumption_per_day = grains_per_person_per_day * number_people

  try:
    survival_days = number_rice_grains // total_consumption_per_day 
    return survival_days
  except ZeroDivisionError:
    return 0 

# Example usage:
total_rice_grains = 100000  # the number of rice grains
num_students = 30  # the  number of students in class

survival_days = calculate_survival_days(total_rice_grains, num_students)

print(f"Your class can survive on this amount of rice for {survival_days} days.")
```
